# Examples

### Timeseries

```python
import numpy as np
timestamps = np.arange(1636104609000, 1636104609000 + 40000, 100)
cosinus_series = (np.cos(timestamps / 1000) + 3).tolist()
sinus_series = (np.sin(timestamps / 2000) + 2).tolist()
random_series_1 = (np.random.randint(-1000, 1000, size=(len(timestamps))) / 1000).tolist()
```

```python
import matplotlib.pyplot as plot

plot.plot(timestamps, cosinus_series, label="cos")
plot.plot(timestamps, sinus_series, label="sin")
plot.title('Cosine wave')
plot.legend()
plot.show()
```

```python
series = {"cos": cosinus_series, "sin": sinus_series}
params = {"title": "Cosine Wave"}
DDGraphs.draw_timeseries_graph(timestamps.tolist(), series, params)
```

```python
series = {
    "cos": cosinus_series, 
    "sin": sinus_series, 
    "random_1": random_series_1, 
}
params = {
    "title": "Cosine Wave", 
    "palette": "dd-classic-line", 
    "customHeight": 400,
    "theme": "light",
    "yAxis": {
        "min": -2,
        "max": 5
    }
}
DDGraphs.draw_timeseries_graph(timestamps.tolist(), series, params)
```
